$\textbf{5. Bipartite Graphs}$

This is a bipartite graph because its nodes can be separated into two distinct sets, with every edge connecting a node from one set to a node in the other set, and no edges connecting nodes within the same set.

```python
import matplotlib.pyplot as plt
import networkx as nx

mySeed = 79

node_names = ['A', 'B', 'C', 'D', 'E']
name_map = {i: name for i, name in enumerate(node_names)}
G = nx.bipartite.gnmk_random_graph(2, 3, 6)
G = nx.relabel_nodes(G, name_map)

# Separate nodes into two sets
set1 = {node for node, node_data in G.nodes(data=True) if node_data['bipartite'] == 0}
set2 = set(G) - set1

# Draw bipartite graph
pos = nx.spring_layout(G, seed=mySeed)
nx.draw(G, pos, nodelist=set1, node_color='pink', node_size=900)
nx.draw(G, pos, nodelist=set2, node_color='lime', node_size=900)
nx.draw_networkx_labels(G, pos=pos)

plt.title("Bipartite Graph")
plt.show()
```

$\textbf{8. Handshake Theorem}$

It states that in any finite undirected graph, the sum of the degrees of all vertices is twice the number of edges.
Equation:

$∑​deg(v)=2e$, such that deg(v) = degree of a vertex, and e = total number of edges

Example from the Previous Graph:

$Edges(e) = 8$

$TotalDegrees(t) = 16$

Thus,

$t=2(8)=16$

$\textbf{12. Euler's Circuit}$

This graph qualifies as an Euler's Circuit because it fulfills the following conditions:

- All vertices in the graph have an even degree.

- The graph is connected.

```python
import matplotlib.pyplot as plt
import networkx as nx

node_names = ['A', 'B', 'C', 'D', 'E', 'F']
edge_list = [("A", "B"), ("B", "C"), ("C", "D"), ("D", "E"), ("E", "F"), ("F", "A"), ("A", "C"), ("C", "E"), ("A", "E")]

# Create the graph
G = nx.Graph()
G.add_nodes_from(node_names)
G.add_edges_from(edge_list)

# Calculate the Eulerian circuit
euler_circuit = list(nx.eulerian_circuit(G))

# Generate positions for the nodes
pos = nx.spring_layout(G, seed=345)

# Draw the graph with nodes and labels
nx.draw(G, pos=pos, with_labels=True, node_color='lime', node_size=900)

# Draw red arrows for the Eulerian circuit
for u, v in euler_circuit:
    x0, y0 = pos[u]
    x1, y1 = pos[v]
    dx = x1 - x0
    dy = y1 - y0
    mid_x = (x0 + x1) / 2
    mid_y = (y0 + y1) / 2
    arrow_length_factor = 0.2
    plt.arrow(mid_x, mid_y, dx * arrow_length_factor, dy * arrow_length_factor, head_width=0.02, head_length=0.06, fc='salmon', ec='salmon')

plt.title("Euler's Circuit")
plt.show()
```

$\textbf{14. Hamiltonian Circuits}$

The graph depicted below represents a Hamiltonian circuit, where the path traverses every vertex exactly once without revisiting the same vertex.


```python
import matplotlib.pyplot as plt
import networkx as nx

def hamilton(G):
    F = [(G,[list(G.nodes())[0]])]
    n = G.number_of_nodes()
    while F:
        graph,path = F.pop()
        confs = []
        neighbors = (node for node in graph.neighbors(path[-1])
                     if node != path[-1]) #exclude self loops
        for neighbor in neighbors:
            conf_p = path[:]
            conf_p.append(neighbor)
            conf_g = nx.Graph(graph)
            conf_g.remove_node(path[-1])
            confs.append((conf_g,conf_p))
        for g,p in confs:
            if len(p)==n:
                return p
            else:
                F.append((g,p))
    return None

node_names = ['A', 'B', 'C', 'D']
edge_list = [("A", "B"), ("B", "C"), ("C", "D"), ("D", "A"), ("A", "C"), ("B", "D")]

G = nx.Graph()
G.add_nodes_from(node_names)
G.add_edges_from(edge_list)

hamiltonian_c = list(hamilton(G))

# Visualize the graph
# Visualize the graph
pos = nx.spring_layout(G, seed=65)
nx.draw(G, pos=pos, with_labels=True, node_color='lime', node_size=900)

# Draw edges from Hamiltonian circuit
for i in range(len(hamiltonian_c) - 1):
    u, v = hamiltonian_c[i], hamiltonian_c[i + 1]
    nx.draw_networkx_edges(G, pos=pos, edgelist=[(u, v)], arrows=True, arrowsize=20, edge_color='red')

u, v = hamiltonian_c[-1], hamiltonian_c[0]
nx.draw_networkx_edges(G, pos=pos, edgelist=[(u, v)], arrows=True, arrowsize=20, edge_color='red')

plt.title("Hamiltonian Circuit")
plt.show()
```

$\textbf{14. Travelling Salesman Problem}$

The graph depicted below represents a Hamiltonian circuit, where the path traverses every vertex exactly once without revisiting the same vertex.


```python
import matplotlib.pyplot as plt
import networkx as nx

# Define the nodes and edges
nodes = ['A', 'B', 'C', 'D', 'E']

edge_labels = {
    ('A', 'B'): 10,
    ('A', 'C'): 20,
    ('A', 'D'): 30,
    ('A', 'E'): 40,
    ('B', 'C'): 50,
    ('B', 'D'): 60,
    ('B', 'E'): 70,
    ('C', 'D'): 80,
    ('C', 'E'): 90,
    ('D', 'E'): 100,
    ('E', 'D'): 5
}

# Define Hamilton circuits
hamilton_circuits = [
    ['A', 'B', 'C', 'D', 'E', 'A'],
    ['A', 'B', 'D', 'E', 'C', 'A'],
    ['A', 'B', 'E', 'D', 'C', 'A'],
    ['A', 'E', 'D', 'C', 'B', 'A'],
    ['A', 'E', 'D', 'B', 'C', 'A']
]

# Plotting
G = nx.Graph()
G.add_nodes_from(nodes)

# Add edges with formatted labels
for edge, label in edge_labels.items():
    G.add_edge(edge[0], edge[1], weight=label)

pos = nx.circular_layout(G)

nx.draw_networkx_edge_labels(
    G, pos,
    edge_labels=edge_labels,
    font_color='red'
)

nx.draw(G, pos=pos, with_labels=True, node_color='lime', node_size=900)

# Highlighting the specified Hamilton circuit (Route 5)
for j in range(len(hamilton_circuits[4]) - 1):
    nx.draw_networkx_edges(
        G, pos, edgelist=[(hamilton_circuits[4][j], hamilton_circuits[4][j + 1])],
        edge_color='red', width=2
    )

# Calculating and displaying the weight of the route
weight = sum(G[hamilton_circuits[4][j]][hamilton_circuits[4][j + 1]]['weight'] for j in range(len(hamilton_circuits[4]) - 1))
plt.text(0.5, -0.15, 'Weight: {}'.format(weight), horizontalalignment='center', verticalalignment='center',
         transform=plt.gca().transAxes)

# Show the plot
plt.show()
```

Among the five potential routes displayed for the graph, the most optimal and shortest path is:


A->E->D->B->C->A = 40 + 5 +60 + 50 + 20 = 175
